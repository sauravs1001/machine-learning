### Method Chaining

let's have a simple example here. So first of all let's try to calculate the mean value in all columns and we can actually do this with the mean() method, so here we have the mean values in our columns and for instance the the survival rate is 38.3% and we can also sort of those values from high to low or from low to high and we can simply chain another method.
That's the sort values method and let's call the method and the by default the sort_values() method sort all elements from low to high so we're starting with the lowest mean in the parents and children column
lowest to the highest mean value in the Fair column. And finally we can also only have a look at the the first two elements and the it's no surprise that we can chain here the head() method and we can pass 2 to get to the first two elements.

And again this concept is called Method chaining as we chain the methods one after another.

```python
# titanic.mean()
# titanic.mean().sort_values()

titanic.mean().sort_values().head(2)
```

### Indexing/Slicing Rows and Columns with loc operator

```python
# To display all the medals of Michael Phelps
summer.loc["PHELPS, Michael", "Medal"]
```

```python
# To get more than 1 column we can simply pass a list of column names on the right hand side of comma

summer.loc["PHELPS, Michael", ["Event", "Medal"]]
```

```python
# To select all the rows and required columns with the .loc operator

summer.loc[: , ["Event", "Medal"]]
```

```python
summer.head(10)
```

Note:
- We can also perform slicing so from two and actually labeled by slicing with text data on the indexes is not the most relevant case and practice. So there is hardly any reason why you should slice from Michael Phelps to Carl Lewis. But this is a good example to show you some pitfalls when slicing an index of duplicate entries and in practice a label by slicing is heavily used with the date time data on the index a so-called date and time index and you will see this in later sections of this course. But for the time being let's assume that we want to get all rows from the very beginning to including the row with Spiridon CHASAPIS athlete. And actually this is pretty straight forward with the loc operator

```python
# To display all the record from beginning till "CHASAPIS, Spiridon" athlete

summer.loc[:"CHASAPIS, Spiridon"]
```

Note:
- It's important to know the difference between iloc operator and loc operator
- In iloc operator the left boundary is included the right boundary is excluded
- In loc operator both the left and right boundaries are included


```python
# To display all the record from beginning till "PHELPS, Michael" athlete

summer.loc[:"PHELPS, Michael"]
```

Note:
- In the above example PHELPS, Michael record is a duplicate on our index therefore we cannot perform slicing operations 

```python
# We will also get the same error message if we try to slice from "PHELPS, Michael" till last row

summer.loc["PHELPS, Michael": ]
```

```python
# If Labels are unique then we can slice information from a dataframe using loc operator and labelled index

summer.loc["DRIVAS, Dimitrios":"BLAKE, Arthur"] # Both boundaries are included in loc operator
```

```python
# To get columns info while slicing using left and right bound. again both boundaries inclusive due to loc operator

summer.loc["DRIVAS, Dimitrios":"BLAKE, Arthur", "City": "Discipline"]
```

```python
# Try to display information using labels that does not exist in the dataframe

summer.loc[["PHELPS, Michael", "DUCK, Donald"]] # Using a list to pass 2 row labels
```

### Indexing and Slicing with reindex() method

 - Reindex is frequently used when working with time series data where we have a so called date time index.
 - But we can use reindex() method for other purposes as well.
 - One advantage of reindex() method is that we can pass the row labels or column labels to reindex that are not available in the dataframe.
 - Previously we saw that missing labels doesnot work with .loc operator.

```python
summer1 = pd.read_csv("summer.csv")
```

```python
summer1
```

```python
# With .loc operator we can also use range index along with label index
# But we should not mix range index and labelled index

summer1.loc[[0, 5, 30000], ["Athlete", "Medal"]]
```

```python
# Using reindex method to work with missing row index, in this case it is a range index
# So we will get an additional 40000 indexed row with NaN values

summer1.reindex(index=[0, 5, 30000, 40000], columns = ["Athlete", "Medal"])
```

```python
# We can also add another column label that is missing from the dataframe

summer1.reindex(index=[0, 5, 30000, 40000], columns = ["Athlete", "Medal", "Age"]) # Age column is extra
```

 - Reindex method gives us some flexibility but there are also limitations. 
 - Reindex method doesn't work in case we do not have unique column index or row index.
 - reindex() method always take the argument values as a list, even if it's only one element. 

```python
summer2 = pd.read_csv("summer.csv", index_col="Athlete")
```

```python
summer2
```

```python
summer2.reindex(index=["PHELPS, Michael"], columns=["Medal", "Age"]) # Here row index is having duplicate entries
```
